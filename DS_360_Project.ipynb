1. What is the average transaction amount ($) across different store types, and how does it vary by season?

```python
avg_transaction_by_store_season = df.groupby(['Store_Type', 'Season'])['Amount($)'].mean().reset_index()
avg_transaction_by_store_season
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_transaction_by_store_season, x='Season', y='Amount($)', hue='Store_Type') # Changed y to 'Amount($)'
plt.title("Average Transaction Amount by Store Type and Season")
plt.ylabel("Average Transaction Amount ($)")
plt.xlabel("Season")
plt.legend(title="Store Type")
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

2. Which payment method is most commonly used in high-value transactions (above the average transaction amount), and how does it differ across cities?

```python
average_transaction_amount = df['Amount($)'].mean()
high_value_transactions = df[df['Amount($)'] > average_transaction_amount]

payment_method_by_city = high_value_transactions.groupby(['City', 'Payment_Method']).size().reset_index(name='Count')
most_common_payment_method = payment_method_by_city.loc[payment_method_by_city.groupby('City')['Count'].idxmax()]
print("Most Common Payment Method in High-Value Transactions Across Cities:")
print(most_common_payment_method)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=most_common_payment_method, x='City', y='Count', hue='Payment_Method')
plt.title("Most Common Payment Method for High-Value Transactions Across Cities")
plt.ylabel("Transaction Count")
plt.xlabel("City")
plt.legend(title="Payment Method")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

3. How do the sales amounts in transactions with discounts compare to those without discounts, and what trends can be observed over the month?

```python
sales_with_discounts = df[df['Discount_Applied'] == True]['Amount($)'].sum()
sales_without_discounts = df[df['Discount_Applied'] == False]['Amount($)'].sum()

print("Total Sales in Transactions with Discounts:", sales_with_discounts)
print("Total Sales in Transactions without Discounts:", sales_without_discounts)

df['Month'] = pd.to_datetime(df['Date']).dt.month
monthly_sales = df.groupby(['Month', 'Discount_Applied'])['Amount($)'].sum().reset_index()
print("Monthly Sales with and without Discounts:")
print(monthly_sales)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_sales, x='Month', y='Amount($)', hue='Discount_Applied')
plt.title("Monthly Sales: With Discounts vs Without Discounts")
plt.ylabel("Sales Amount ($)")
plt.xlabel("Month")
plt.legend(title="Discount Applied", labels=["No", "Yes"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

4. What are the top three cities with the highest average number of items per transaction, and how do their sales amounts vary across seasons?

```python
avg_items_by_city = df.groupby('City')['Total_Items'].mean().reset_index()
top_3_cities = avg_items_by_city.nlargest(3, 'Total_Items')

sales_by_city_season = df[df['City'].isin(top_3_cities['City'])].groupby(['City', 'Season'])['Amount($)'].sum().reset_index()

print("Top 3 Cities by Average Items Per Transaction:")
print(top_3_cities)
print("Sales Amounts for Top 3 Cities Across Seasons:")
print(sales_by_city_season)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=top_3_cities, x='City', y='Total_Items', palette='Blues_d')
plt.title("Top 3 Cities by Average Items Per Transaction")
plt.ylabel("Average Items Per Transaction")
plt.xlabel("City")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(data=sales_by_city_season, x='Season', y='Amount($)', hue='City', marker='o')
plt.title("Sales Amounts for Top 3 Cities Across Seasons")
plt.ylabel("Sales Amount ($)")
plt.xlabel("Season")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

5. How effective are different promotions in driving higher transaction amounts, and which promotion type performs best in each season?

```python
promotion_effectiveness = df.groupby(['Promotion', 'Season'])['Amount($)'].mean().reset_index()
promotion_effectiveness.columns = ['Promotion', 'Season', 'Average_Transaction_Amount']

best_promotion_by_season = promotion_effectiveness.loc[promotion_effectiveness.groupby('Season')['Average_Transaction_Amount'].idxmax()]

print("Effectiveness of Promotions:")
print(promotion_effectiveness)
print("Best Promotion by Season:")
print(best_promotion_by_season)
```

```python
pivot = promotion_effectiveness.pivot(index="Promotion", columns="Season", values="Average_Transaction_Amount")
plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("Effectiveness of Promotions by Season")
plt.ylabel("Promotion")
plt.xlabel("Season")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(data=best_promotion_by_season, x='Season', y='Average_Transaction_Amount', hue='Promotion')
plt.title("Best Promotion by Season")
plt.ylabel("Average Transaction Amount ($)")
plt.xlabel("Season")
plt.legend(title="Promotion")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
